## Clean Post Text

### Subtask:
Remove stopwords, punctuation, and special symbols from the 'post text' column to prepare it for analysis.


### Subtask:
Remove stopwords, punctuation, and special symbols from the 'post text' column to prepare it for analysis.

#### Instructions
1. Import the `re` module for regular expressions and `string` module for punctuation.
2. Import `nltk` and download the 'stopwords' corpus if it hasn't been downloaded already.
3. Define a function, for example `clean_text`, that takes a string as input and performs the following operations:
   a. Convert the text to lowercase.
   b. Remove HTML tags (e.g., using `re.sub(r'<.*?>', '', text)`).
   c. Remove punctuation (e.g., using `str.translate` with `str.maketrans('', '', string.punctuation)`).
   d. Remove numbers (e.g., using `re.sub(r'\d+', '', text)`).
   e. Remove stopwords (e.g., by splitting the text into words and filtering out common English stopwords).
   f. Remove extra whitespace.
4. Apply this `clean_text` function to the 'post_text' column of the DataFrame `df`, updating the column with the cleaned text.

**Reasoning**:
Now that the instructions are laid out, I will implement the text cleaning steps. This involves importing necessary libraries, downloading NLTK stopwords, defining a function to clean text by lowercasing, removing HTML tags, punctuation, numbers, and stopwords, and finally applying this function to the 'post_text' column.



```python
import re
import string
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    if not isinstance(text, str): # Handle non-string types, e.g., NaN
        return text
    text = text.lower() # Convert to lowercase
    text = re.sub(r'<.*?>', '', text) # Remove HTML tags
    text = text.translate(str.maketrans('', '', string.punctuation)) # Remove punctuation
    text = re.sub(r'\d+', '', text) # Remove numbers

    # Remove stopwords and extra whitespace
    text = ' '.join([word for word in text.split() if word not in stop_words])
    text = re.sub(r'\s+', ' ', text).strip() # Remove extra whitespace
    return text

df['post_text'] = df['post_text'].apply(clean_text)
print("Cleaned 'post_text' column preview:")
print(df[['post_text']].head())
```

**Reasoning**:
The previous step successfully cleaned the 'post_text' column. Following the task description, the next step is to handle missing values in the 'likes' and 'shares' columns.



```python
print("Missing values before handling:")
print(df[['likes', 'shares']].isnull().sum())

# Fill missing 'likes' with the median
df['likes'].fillna(df['likes'].median(), inplace=True)

# Fill missing 'shares' with the median
df['shares'].fillna(df['shares'].median(), inplace=True)

print("\nMissing values after handling:")
print(df[['likes', 'shares']].isnull().sum())

print("\nDataFrame head after handling missing values:")
print(df.head())
```

**Reasoning**:
The previous code produced `FutureWarning` due to the use of `inplace=True` with `fillna` on chained assignments. To address this, I will update the code to assign the result of the `fillna` operation back to the DataFrame column directly, which is the recommended practice to avoid such warnings and ensure proper behavior in future pandas versions.



```python
print("Missing values before handling:")
print(df[['likes', 'shares']].isnull().sum())

# Fill missing 'likes' with the median
df['likes'] = df['likes'].fillna(df['likes'].median())

# Fill missing 'shares' with the median
df['shares'] = df['shares'].fillna(df['shares'].median())

print("\nMissing values after handling:")
print(df[['likes', 'shares']].isnull().sum())

print("\nDataFrame head after handling missing values:")
print(df.head())
```

**Reasoning**:
The previous steps successfully handled missing values and cleaned the 'post_text' column. The next logical step, according to the overall task, is to process the 'timestamp' column by converting it to datetime objects and extracting the 'hour' and 'weekday' components.



```python
print("Timestamp column before processing:")
print(df[['timestamp']].head())
print(df['timestamp'].dtype)

# Convert 'timestamp' to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Extract 'hour' and 'weekday'
df['hour'] = df['timestamp'].dt.hour
df['weekday'] = df['timestamp'].dt.day_name()

print("\nTimestamp column after processing:")
print(df[['timestamp', 'hour', 'weekday']].head())
print(df['timestamp'].dtype)
```
